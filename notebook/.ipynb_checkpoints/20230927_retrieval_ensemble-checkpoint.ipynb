```python

def recallk(y_true, y_pred, k=5):
    """
    y_true: (n_samples)
    y_pred: (n_samples, n_classes)
    """
    map_score = 0
    for i in range(len(y_true)):
        y_true_ = y_true[i]
        y_pred_ = y_pred[i]
        score = 0
        for j in range(min(k, len(y_pred_))):
            if y_pred_[j] == y_true_:
                score += 1
                break
        map_score += score
    map_score /= len(y_true)
    return map_score
```

```python
df2 = pd.read_parquet("../output/context_pipeline/stage1/exp010.py/20230926083926_gte-base_wikiall_without_sep_targetprompt_and_choice_without_sep_token_length120_stride_sentence4_drop_categoryTrue_only_a_sep_abcde//train.parquet")
```

```python
df2["first_id"] = [x[0] for x in df2["searched_ids"].values]
```

```python
df_concat = []
for wiki_id, w_df in df2.groupby("wiki_id"):
    df_concat.append({
        "wiki_id": str(wiki_id),
        "searched_ids": w_df["first_id"].values.astype(str)
    })
df_concat = pd.DataFrame(df_concat)
```

```python
recallk(df_concat["wiki_id"].values, df_concat["searched_ids"].values, k=5)
```

```python
df_concat["correct"] = [str(wid) in sid for wid, sid in zip(df_concat["wiki_id"].values, df_concat["searched_ids"].values)]
```
