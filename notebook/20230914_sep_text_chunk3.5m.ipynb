```python
import pandas as pd
import tqdm
import glob
import os
```

```python
files = glob.glob("../data/wikipedia/*.parquet")
files = [f for f in files if "all" not in f]
```

```python
def output_seperated_files(token_length, stride):    
    output_dir = f"../data/wikipedia/sep_chunk/token_length{token_length}_stride{stride}"
    os.makedirs(output_dir, exist_ok=True)
    chunk = 1
    df_wiki_sep = []
    for file in files:
        df_wiki = pd.read_parquet(file)
        df_wiki = df_wiki.drop("categories", axis=1)
        for i in tqdm.tqdm(range(len(df_wiki))):
            series = df_wiki.iloc[i]
            text = series["text"]

            text_space = text.split(" ")
            for i in range(0, len(text_space) + stride, stride):
                series_ = series.copy()
                series_["text"] = " ".join(text_space[i:i+token_length])
                df_wiki_sep.append(series_)
            if len(df_wiki_sep) > 3_500_000:
                print(f"chunk {chunk}")
                df_wiki_sep = pd.DataFrame(df_wiki_sep)
                df_wiki_sep.to_parquet(f"{output_dir}/chunk{chunk}.parquet")
                df_wiki_sep = []
                chunk += 1
    df_wiki_sep = pd.DataFrame(df_wiki_sep)
    df_wiki_sep.to_parquet(f"{output_dir}/chunk{chunk}.parquet")
    
```

```python
output_seperated_files(token_length=100, stride=75)
```
