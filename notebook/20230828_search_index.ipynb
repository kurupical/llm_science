```python
import pandas as pd
import glob
import tqdm
```

```python
df_train = pd.read_csv("../output/searched_index.csv")
```

```python
df_wiki = pd.concat([pd.read_parquet(f) for f in tqdm.tqdm(glob.glob("../data/wikipedia/*.parquet"))])
```

```python
df_wiki["text"] = df_wiki["title"] + "\n" + df_wiki["text"]
wiki_dict = df_wiki.set_index("id")["text"].to_dict()
```

```python
for i in range(100):
    series = df_train.iloc[i]
    print("\n\n=================================")
    print(f"original_text: {series['prompt']}")
    
    print("----------------------")
    print(f"answer: {series['original_text']}")
    
    searched_indice = series["searched_index"]
    searched_indice = [int(x) for x in searched_indice.replace("[", "").replace("]", "").replace(" ", "").replace("'", "").split(",")]
    for j, idx in enumerate(searched_indice[:3]):
        print("----")
        if series["wiki_id"] == idx:
            print(f"●similar {j}: {wiki_dict[str(idx)][:1000]}")            
        else:
            print(f"similar {j}: {wiki_dict[str(idx)][:1000]}")
```
